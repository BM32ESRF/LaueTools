```python
def Ge_attenuation(E): # in micron
    return np.where(E<11., (66-10)/6*(E-5)+10, (60-10)/11*(E-11)+10)

erange = np.arange(5,22,.1)
fig, ax =plt.subplots()
ax.scatter(erange,Ge_attenuation(erange))
```

```python
# correction in y pixel 
-Ge_attenuation(erange)/1000./0.079142
```

```python
X, Y = Xexp, Yexp -0.*((Ge_attenuation(E)+0)/1000./0.079142)
In = Intensity
outputprefixfilename = 'depthcorrected_Ge'

Data_array = np.array([X,Y,In]).T
IOLT.writefile_Peaklist(outputprefixfilename, Data_array, overwrite=1,
                                                        initialfilename=filename,
                                                        comments=None,
                                                        dirname=folder,
                                                        verbose=0)
```
